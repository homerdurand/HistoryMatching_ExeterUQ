* You can select the number of variables X you want to keep (var *n_k*)
* Data should be with format c(inputs, Noise, outputs) and named *tData*

```r
library(BBmisc)#For normalization
library(comprehenr)#For comprehension loop
library(caret)

N = 40 

#How many variables you want to keep
n_k = 36
columns_to_keep = c()
columns_to_keep = to_vec(for(i in seq(1, n_k)) c(columns_to_keep, seq(i, 180, 36)))

#Load and normalize inputs (parameters)
inputs <- read.csv("/home/homer/Documents/Locean/Data/inputs.csv", header = TRUE)
normalizer <- preProcess(inputs, method=c("range"), rangeBounds = c(-1, 1))
inputs <- predict(normalizer, inputs)


#Load outputs and select variables you want to keep
outputs <- read.csv("/home/homer/Documents/Locean/Data/outputs.csv", header = TRUE)
outputs_names  = c(rep('X', 36), rep('Y_mean', 36), rep('X_sq', 36), rep('XY_mean', 36), rep('Y_mean_sq', 36))
names(outputs) <- outputs_names
outputs <- outputs[, columns_to_keep]
                         
#Add some noise (here no noise added)
noise <- rnorm(N, 0, 0)
tData <- cbind(inputs, noise, outputs)
names(tData)[names(tData) == "noise"] <- "Noise"
                         
head(tData)
```

### Training Emulators

```r
TestEm <- BuildNewEmulators(tData, HowManyEmulators = length(outputs), meanFun = "fitted", additionalVariables = names(tData)[1:4]) # 
```

## Testing Emulators

## History Matching
* Observations discrepancy variances in *tDisc* (here sets to 0)
* Observations errors in *tObsErr* (here set to 0)

```r
tDisc <- rep(0.5, length(outputs))
tObs <- read.csv("/home/homer/Documents/Locean/Data/observations.csv", header = FALSE)
tObs <-  as.vector(t(tObs[, columns_to_keep]))
tObsErr <- rep(0, length(outputs))
```

Modifier fonction *ImplausibilityMOGP()* avec GPytorch 

```r
sample_size <- 10000
nparam <- length(names(TestEm$fitting.elements$Design))
Xp <- as.data.frame(2*randomLHS(sample_size, nparam)-1)
names(Xp) <- names(TestEm$fitting.elements$Design)
system.time(Timps <- ImplausibilityMOGP(NewData=Xp, Emulator=TestEm, Discrepancy=tDisc, Obs=tObs, ObsErr=tObsErr))
```

```r
ImpData_wave1 = cbind(Xp, Timps)
```

```r
VarNames <- names(Xp)
valmax = 0 #how many outputs can be above the implausibility cut off?
cutoff_vec <- 3 #the implausibility cut off
param.def = data.frame(h=c(1), F=c(10),c=c(10),b=c(10)) #the default parameters of the model (on [-1,1])
param.defaults.norm = predict(normalizer, param.def)
print(param.defaults.norm)
```

```r
ImpListM1 = CreateImpList(whichVars = 1:nparam, VarNames=VarNames, ImpData=ImpData_wave1, nEms=TestEm$mogp$n_emulators, whichMax=valmax+1)
NROY1 <- which(rowSums(Timps <= cutoff_vec[1]) >= TestEm$mogp$n_emulators -valmax)
length(NROY1)/dim(Xp)[1]
```

```r
imp.layoutm11(ImpListM1,VarNames,VariableDensity=FALSE,newPDF=FALSE,the.title=paste("InputSpace_wave",WAVEN,".pdf",sep=""),newPNG=FALSE,newJPEG=FALSE,newEPS=FALSE,Points=matrix(param.defaults.norm,ncol=nparam))
mtext(paste("Remaining space:",length(NROY1)/dim(Xp)[1],sep=""), side=1)
```

## Test on number of variables

```r
computeNROY <- function(emulator, sample_size = 10000, valmax = 0, cutoff_vec = 3, disc = tDisc, obs = tObs, obserr = tObsErr)
{
    nparam <- length(names(TestEm$fitting.elements$Design))
    Xp <- as.data.frame(2*randomLHS(sample_size, nparam)-1)
    names(Xp) <- names(emulator$fitting.elements$Design)
    Timps <- ImplausibilityMOGP(NewData=Xp, Emulator=emulator, Discrepancy=disc, Obs=obs, ObsErr=obs)
    NROY1 <- which(rowSums(Timps <= cutoff_vec[1]) >= emulator$mogp$n_emulators -valmax)
    return(length(NROY1)/dim(Xp)[1])
}
```

```r
#Set discrepency and obs errors
tDisc <- rep(0, length(outputs))
tObsErr <- rep(0, length(outputs))

nroyspace = c()
for (n_k in 1:10)
{
    #rebuild data with correct n_k
    columns_to_keep = sapply(1:n_k, function(k) seq(k, 180, 36))
    outputs <- read.csv("/home/homer/Documents/Locean/Data/outputs.csv", header = TRUE)
    outputs <- outputs[, columns_to_keep]
    tData <- cbind(inputs, noise, outputs)
    names(tData)[names(tData) == "noise"] <- "Noise"
    
    #Select rows of tObs you want to keep 
    tObs <- read.csv("/home/homer/Documents/Locean/Data/observations.csv", header = FALSE)
    tObs <-  as.vector(t(tObs[, columns_to_keep]))
    
    myEm <- BuildNewEmulators(tData, HowManyEmulators = length(outputs), meanFun = "fitted", additionalVariables = names(tData)[1:4]) # 
    nroyspace[n_k] <- computeNROY(myEm)
}
```

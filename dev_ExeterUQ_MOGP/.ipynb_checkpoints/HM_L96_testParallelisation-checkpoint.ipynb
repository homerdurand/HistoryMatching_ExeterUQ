## Testing Emulators

```r
LOO.score <- function(emulator){
    #Compute pourcentage of targets which not fall in 2*sigma interval
    scores <- c()
    for(which.emulator in 1:emulator$mogp$n_emulators){
        fit.loo <- virtual.LOO_MOGP(mogp.emulator=TestEm$mogp$emulators[[which.emulator]], lm.emulator=TestEm$fitting.elements$lm.object[[which.emulator]])
        errors = sum(emulator$mogp$emulators[[which.emulator]]$targets < fit.loo$`lower quantile` | emulator$mogp$emulators[[which.emulator]]$targets > fit.loo$`upper quantile`)
        score = errors/length(emulator$mogp$emulators[[which.emulator]]$targets)
        scores[which.emulator] = score
    }
    return(c(mean = mean(scores), min = max(scores), scores = data.frame(scores)))
}
```

```r
loo = LOO.score(TestEm)
df = data.frame(loo)
library(ggplot2)
# Graphe basique
p <- ggplot(df, aes(x=scores.scores)) + geom_histogram(color="darkblue", fill="lightblue", binwidth = 0.025,) 
p + geom_vline(aes(xintercept=mean(scores.scores)), color="blue", linetype="dashed", size=2)
```

```r
tDisc <- rep(0, length(outputs))
tObs <- read.csv("/home/homer/Documents/Locean/Data/observations.csv", header = FALSE)
tObs <-  as.vector(t(tObs[, columns_to_keep]))
tObsErr <- rep(0, length(outputs))
#varObsErr <- var(tObs)
#pvar = varObsErr*0.05
#tObsErr = abs(rnorm(180, 0, pvar))
```

```r
sample_size <- 20000
nparam <- length(names(TestEm$fitting.elements$Design))
Xp <- as.data.frame(2*randomLHS(sample_size, nparam)-1)
names(Xp) <- names(TestEm$fitting.elements$Design)
system.time(Timps <- ImplausibilityMOGP(NewData=Xp, Emulator=TestEm, Discrepancy=tDisc, Obs=tObs, ObsErr=tObsErr))
```

```r
ImpData_wave1 = cbind(Xp, Timps)
```

```r
VarNames <- names(Xp)
valmax = 0 #how many outputs can be above the implausibility cut off?
cutoff_vec <- 3 #the implausibility cut off
param.def = data.frame(h=c(1), F=c(10),c=c(10),b=c(10)) #the default parameters of the model (on [-1,1])
param.defaults.norm = predict(normalizer, param.def)
print(param.defaults.norm)
```

```r
ImpListM1 = CreateImpList(whichVars = 1:nparam, VarNames=VarNames, ImpData=ImpData_wave1, nEms=TestEm$mogp$n_emulators, whichMax=valmax+1)
NROY1 <- which(rowSums(Timps <= cutoff_vec[1]) >= TestEm$mogp$n_emulators -valmax)
length(NROY1)/dim(Xp)[1]
```

## Test on noise

```r
computeNROY <- function(emulator, sample_size = 10000, valmax = 0, cutoff_vec = 3, disc = tDisc, obs = tObs, obserr = tObsErr)
{
    nparam <- length(names(TestEm$fitting.elements$Design))
    Xp <- as.data.frame(2*randomLHS(sample_size, nparam)-1)
    names(Xp) <- names(emulator$fitting.elements$Design)
    Timps <- ImplausibilityMOGP(NewData=Xp, Emulator=TestEm, Discrepancy=tDisc, Obs=tObs, ObsErr=tObsErr)
    NROY1 <- which(rowSums(Timps <= cutoff_vec[1]) >= TestEm$mogp$n_emulators -valmax)
    length(NROY1)/dim(Xp)[1]
}

my_log <- function(x) log(x)/35 +0.05
```

```r
nroyspace = data.frame(x=NA, y=NA)
varObsErr <- var(tObs)
for(i in 1:50){
    pvar <- varObsErr*(i/100)
    nroy = matrix()
    for (j in 1:5){
        tObsErr = abs(rnorm(180, 0, pvar))
        y<- computeNROY(TestEm, sample_size = 1000, valmax = 0, cutoff_vec = 3, disc = tDisc, obs = tObs, obserr = tObsErr)
        x <- i
        nroyspace <- rbind(nroyspace, c(x, y))
    }
}
```

```r
library(ggplot2)
library(dplyr)

nroyspace.summary <- nroyspace %>% group_by(x) %>%
    summarize(ymin = min(y),
              ymax = max(y),
              ymean = mean(y))

ggplot(nroyspace.summary, aes(x = x, y = ymean)) +
    geom_point(size = 2) +
    geom_errorbar(aes(ymin = ymin, ymax = ymax)) + 
    labs(title= "NROY space regarding noise as percentage of observations variance",
                      y="NROY space", x = "noise variance (as percentage of obs variance)") +
    stat_function(fun = my_log, color="red") 
```

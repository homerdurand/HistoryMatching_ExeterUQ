* You can select the number of variables X you want to keep (var *n_k*)
* Data should be with format c(inputs, Noise, outputs) and named *tData*

```r
library(BBmisc)#For normalization
library(comprehenr)#For comprehension loop
library(caret)

#How many variables you want to keep
n_k = 36
columns_to_keep = c()
columns_to_keep = to_vec(for(i in seq(1, n_k)) c(columns_to_keep, seq(i, 180, 36)))

#Load and normalize inputs (parameters)
inputs <- read.csv("/home/homer/Documents/Locean/Data/clean_inputs_40.csv", header = TRUE)
inputs_names <- c('h', 'F', 'c', 'b')
names(inputs) <- inputs_names
my_bounds <- data.frame(list(c(-2, 2), c(-20, 20), c(0, 20), c(-20, 20)))
inputs <- rangeScale(inputs, my_bounds, range = c(-1, 1))


#Load outputs and select variables you want to keep
outputs <- read.csv("/home/homer/Documents/Locean/Data/clean_outputs_100_40.csv", header = FALSE)
outputs_names  = c('index', rep('X', 36), rep('Y_mean', 36), rep('X_sq', 36), rep('XY_mean', 36), rep('Y_mean_sq', 36))
names(outputs) <- outputs_names
outputs <- outputs[2:nrow(outputs), ]
outputs$index <- NULL
outputs <- outputs[, columns_to_keep]


#Add some noise (here no noise added)
N = nrow(inputs) 
noise <- rnorm(N, 0, 0)
tData <- cbind(inputs, noise, outputs)
names(tData)[names(tData) == "noise"] <- "Noise"
         
head(tData)
```

### Training Emulators

```r
em_wave1 <- BuildNewEmulators(tData, HowManyEmulators = 180, meanFun = "fitted", additionalVariables = names(tData)[1:4]) # 
```

## History Matching
* Observations discrepancy variances in *tDisc* (here set to 0)
* Observations errors in *tObsErr* (here set to 0)

```r
tDisc <- rep(0, length(outputs))
tObs <- read.csv("/home/homer/Documents/Locean/Data/observations.csv", header = FALSE)
tObs <-  as.vector(t(tObs[, columns_to_keep]))
tObsErr <- rep(0, length(outputs))
```

Probleme de dimension : un sampling de 1 000 000 pts ne 
découpe chaque plan qu'en une grille d'environ 32x32

un sampling de 10 000 pts ne 
découpe chaque plan qu'en une grille d'environ 10x10


```r
sample_size <- 160000
nparam <- length(names(em_wave1$fitting.elements$Design))
Xp <- as.data.frame(2*randomLHS(sample_size, nparam)-1)
names(Xp) <- names(em_wave1$fitting.elements$Design)
system.time(Timps <- ImplausibilityMOGP(NewData=Xp, Emulator=em_wave1, Discrepancy=tDisc, Obs=tObs, ObsErr=tObsErr))
```

```r
ImpData_wave1 = cbind(Xp, Timps)
```

```r
VarNames <- names(Xp)
valmax = 0 #how many outputs can be above the implausibility cut off?
cutoff <- 2 #the implausibility cut off
param.def = data.frame(h=c(1), F=c(10),c=c(10),b=c(10)) #the default parameters of the model (on [-1,1])
param.defaults.norm = rangeScale(param.def, my_bounds, range(-1,1))
print(param.defaults.norm)
```

```r
ImpListM1 = CreateImpList(whichVars = 1:nparam, VarNames=VarNames, ImpData=ImpData_wave1, nEms=em_wave1$mogp$n_emulators, whichMax=valmax+1)
NROY1 <- which(rowSums(Timps <= cutoff) >= em_wave1$mogp$n_emulators -valmax)
scoreNROY <- length(NROY1)/dim(Xp)[1]
scoreNROY
```

```r
imp.layoutm11(ImpListM1,VarNames,VariableDensity=FALSE,newPDF=FALSE,the.title=paste("InputSpace_wave",WAVEN,".pdf",sep=""),newPNG=FALSE,newJPEG=FALSE,newEPS=FALSE,Points=matrix(param.defaults.norm,ncol=nparam))
mtext(paste("Remaining space:",length(NROY1)/dim(Xp)[1],sep=""), side=1)
```

```r
Xp2 = Xp[NROY1,]
nrow(Xp2)
```

## Fine Tuning

#### Creating inputs

```r
Nsamples <- 40
cutoff <- 2
```

```r
library(tgp)

ngen <- ceiling(Nsamples*1.2/scoreNROY)
bounds_norm <- data.frame(c(-1,1), c(-1,1), c(-1,1), c(-1,1))
inputs_large <- lhs(ngen, t(bounds_norm))
Timps_temp <- ImplausibilityMOGP(NewData=inputs_large, Emulator=em_wave1, Discrepancy=tDisc, Obs=tObs, ObsErr=tObsErr)
NROY_temp <- which(rowSums(Timps_temp <= cutoff) >= em_wave1$mogp$n_emulators -valmax)
NROY_temp <- sample(NROY_temp, Nsamples)
inputs_wave2 <- as.data.frame(inputs_large[NROY_temp,])
names(inputs_wave2) <- inputs_names
inputs_temp <- rangeUnscale(inputs_wave2, my_bounds, range = c(-1, 1))
nrow(inputs_temp)
```

```r
extract_metrics <- function(p, n_iter) {
    l96 = L96$L96TwoLevel(h=p[1], F=p[2], c=p[3], b=p[4])
    l96$iterate(5)
    l96$erase_history()
    l96$iterate(n_iter)
    return(l96$mean_stats())
}
```

#### Simulating data with L96

```r
n_iter = 15
```

```r
library("future.apply")

plan(multicore) ## Run in parallel on local computer
system.time(temp_outputs <- future_apply(data.matrix(inputs_temp), 1, function(p) {extract_metrics(p,n_iter)}, future.seed=NULL))
```

#### Computing Implausibility


```r
outputs_wave2 = data.frame(t(temp_outputs))
row.names(outputs_wave2) <- NULL
```
